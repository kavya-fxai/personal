```python
import pandas as pd
```

```python
stock_data= pd.read_csv('National_Stock_Exchange_of_India_Ltd.csv')
stock_data
```

To calculate the %K and %D (stochastic indicators) for the stock data, we need to focus on the high, low, and close prices over a specific period- 14 days. This dataset doesnt provide a closing price, so we can use "LTP"- the last traded price.

```python
#remove commas and convert the data to numeric
stock_data['High'] = stock_data['High'].str.replace(',', '').astype(float)
stock_data['Low'] = stock_data['Low'].str.replace(',', '').astype(float)
stock_data['LTP'] = stock_data['LTP'].str.replace(',', '').astype(float)

# Now recalculate the 14-day high and low
stock_data['14-high'] = stock_data['High'].rolling(window=14).max()
stock_data['14-low'] = stock_data['Low'].rolling(window=14).min()
```

```python
#calculate the k
stock_data['%K'] = ((stock_data['LTP'] - stock_data['14-low']) / (stock_data['14-high'] - stock_data['14-low'])) * 100
```

```python
#calculate the d 
stock_data['%D'] = stock_data['%K'].rolling(window=3).mean()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(14, 7))

# Plot %K
plt.plot(stock_data['%K'], label='%K', color='blue')
# Plot %D
plt.plot(stock_data['%D'], label='%D', color='red')

# Add titles and labels
plt.title('Stochastic Oscillator (%K and %D)')
plt.xlabel('Date')
plt.ylabel('Value')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```

%K Line -  This line represents the current closing price's position relative to the highest and lowest prices over the past 14 trading periods. A higher %K value indicates that the current price is near the higher end of the 14-day range, suggesting potential overbought conditions. Conversely, a lower %K value indicates the price is near the lower end of the range, suggesting potential oversold conditions.

%D Line- This line is the 3-day Simple Moving Average (SMA) of the %K line. It smooths out the %K values to provide a clearer signal of potential reversals. When the %K line crosses above the %D line, it can be an indication of a buying opportunity. When the %K line crosses below the %D line, it can suggest a selling opportunity.
